# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel("score.xlsx", index_col = "지원번호")
df
```

```python
# 조건을 이용해 데이터가 조건에 맞는지 확인할 수 있다.
# 키가 185이상 여부 확인
df["키"] >= 185
```

```python
filt = (df["키"]>=185)
df[filt]
#원하는 조건으로 뽑기
```

```python
df[-filt]
#filt 역으로 계산 즉 185미만
```

```python
df[df["키"]>=185] #조건을 직접 넣어 사용도 가능
```

```python
df.loc[df["키"]>=185,"수학"]
#키가 185이상인 데이터의 수학점수
```

```python
df.loc[df["키"]>=185, ["이름","수학","과학"]]
#키가 185이상인 사람들의  이름, 수학 ,과학값 추출
```

## 다양한 조건 

### & 그리고

```python
df.loc[(df["키"] >= 185) & (df["학교"]== "북산고")]

#키가 185 이상인 북산고 학생 데이터
```

### | 또는

```python
df.loc[(df["키"]<=170)|(df["키"]>=200)]
```

### str 함수

```python
filt =df["이름"].str.startswith("송")
df[filt]
# "송" 씨성을 가진 사람
```

```python
filt = df["이름"].str.contains("태")
df[filt]

#태 라는 글자가 들어간 사람 조건
```

```python
df[-filt] #이름에 "태"가 즐어가는 사람을 제외
```

```python
langs =["Python","Java"]
filt=df["SW특기"].isin(langs)
df[filt]
# SW특기가 python이거나 Java인 사람.
```

```python
langs =["python","java"]
filt = df["SW특기"].str.lower().isin(langs)
df[filt]
#SW특기 내부의 값을 소문자로 낮추어 확인한다.
```

```python
filt = df["SW특기"].str.contains("Java", na =True)
df[filt] # true인것은 보여주고, 아닌것은 안보인것.
#Java라는 것을 포함한 애들 부르기
#오류가 뜨는 이유는 NaN 데이터가 있기에 

# na= True #NaN데이터에 대해서 True로 설정
# na= False #NaN데이터에 대해서 False로 설정 
```

```python
filt =df['SW특기'].str.contains("Java", na =False)
df[filt]
```
