## 데이터 채우기 fillna

```python
df.fillna(' ') # na의 데이터를 빈칸으로 채움
```

```python
df.fillna("없음") 
```

```python
import numpy as np
df["학교"]=np.nan  # 학교 데이터를 na로 채운다
df

# numpy는 행렬을 이용해 데이터를 관리하며 연산하는 툴
```

```python
df.fillna("모름") #전체에 대해 넣는 값이다 보니 조심해야 함.
```

```python
df.fillna("모름", inplace = True)
```

### 데이터 제외하기 dropna

```python
df = pd.read_excel("score.xlsx", index_col ="지원번호")
df
# 원본 상태 그대로.
```

```python
df.dropna(inplace = True) # 전체 데이터 중 NaN을 포함하는 데이터 삭제
```

```python
df
```

```python
df = pd.read_excel("score.xlsx", index_col ="지원번호")
df
# 원본 상태 그대로.
```

```python
df.dropna(axis = "index", how = "any") #na가 하나라도 있는 데이터 삭제
```

```python
df.dropna(axis = "columns", how = "any")
```

```python
df["학교"] =np.nan
df
```

```python
df.dropna(axis = "columns", how ="all") # 한 column이 전체 다 NaN일때 삭제.
```
