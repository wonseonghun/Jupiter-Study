# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용 

```python
import pandas as pd
df = pd.read_excel("score.xlsx", index_col = "지원번호")
df
```

```python
df.groupby("학교")
```

```python
df.groupby("학교").get_group("북산고")
#북산고만 가지고 와라.
```

```python
df.groupby("학교").get_group("능남고")
```

```python
df.groupby("학교").mean(numeric_only =True)
#계산가능한 데이터들의 평균
```

```python
df.groupby("학교").size()
```

```python
df.groupby("학교").size()["능남고"]
#학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터의 수
```

```python
df.groupby("학교")["키"].mean(numeric_only =True)
#학교라는 그룹화를 한 후에 키에 대한 평균값을 구한다.
```

```python
df.groupby("학교")[["국어","영어","수학"]].mean(numeric_only =True)
```

```python
df["학년"] = [3,3,2,1,1,3,2,2]
#학년 column 추가
df
```

```python
df.groupby(["학교","학년"]).mean(numeric_only =True)
#학교별, 학년별 평균 데이터
```

```python
df.groupby("학년").mean(numeric_only =True)
#학년 별로 나눌 때 데이터 
```

```python
df.groupby("학년").mean(numeric_only =True).sort_values("키")
#학년 별로 나누고 평균값을 키 값으로 오름차순
```

```python
df.groupby(["학교","학년"]).sum(numeric_only =True)
#numeric_only =True를 안하면 이름도 붙게 된다.
```

```python
df.groupby("학교")[["이름","SW특기"]].count()
#학교 별로 나눈 뒤, 이름을 가진 & SW특기를 가진 사람을 카운트한다.
```

```python
school =df.groupby("학교")
school["학년"].value_counts()
#학교로 그룹화 한 것을 스쿨에 넣고 스쿨에서 각 학년의 개수를 센다.
```

```python
school["학년"].value_counts().loc["북산고"]
# 그 중 북산고의 학년별 학생 수를 알고 싶다.
```

```python
school["학년"].value_counts(normalize=True).loc["북산고"]
#학생 수 데이터를 퍼센트로 비교하여 가져온다.
```
