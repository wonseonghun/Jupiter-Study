# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel("score.xlsx", index_col ="지원번호")
df
```

## Column 수정

```python
df["학교"].replace({"북산고":"상북고", "능남고": "제천고"})  
# 북산고를 상북고로 변경 && 능남고는 제천고로 변경
```

```python
df
# inplace를 하지 않아서 변경 안됨x
```

```python
df["학교"].replace({"북산고" : "상북고","능남고" : "제천고"}, inplace =True)
df
```

```python
df["SW특기"].str.lower()
```

```python
df["SW특기"] = df["SW특기"].str.lower()
df
# 소문자로 변경
```

```python
df["SW특기"] = df["SW특기"].str.upper()
df
# 대문자로 변
```

```python
df["학교"] = df["학교"] + "등학교"
# 학교 데이터 + 등학교
df
```

## Column 추가

```python
df["총합"] = df["국어"]+df["수학"]+df["영어"]+df["사회"]+df["과학"]
df
#총합 column 추가
```

```python
df["결과"]= "Fail" 
#결과 column을 추가하고 전체 데이터를 Fail로 초기화
df
```

```python
df.loc[ df["총합"]>400, "결과"] ="Pass"
#총합이 400보다  큰 row의 결과 데이터에 Pass를 넣어준다.
df
```

## Column 삭제

```python
df.drop(columns =["총합"])
# 총합 Column 을 삭제한다
```

```python
df.drop(columns =["국어","영어","수학"])
#국영수 데이터 삭제
```

## Row 삭제

```python
df.drop(index ="4번")
# 4번 데이터 삭제
```

```python
filt = df["수학"]<80
df[filt].index
#아 인덱스는 이렇게 뽑는구나.
df.drop(index =df[filt].index)
#수학 점수가 80점 미만인 학생 삭제
```
